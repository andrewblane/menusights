```python
import copy
```

```python
def bubblesort(tosort):
    out = copy.copy(tosort)
    flag = 0
    index = 0
    while index+1 < len(tosort):
        #print({"list": tosort, "index": index, "indexvalue": tosort[index] })
        if tosort[index] > tosort[index+1]: 
            tosort[index+1], tosort[index] = tosort[index], tosort[index+1]
            #print("flipped " + str(tosort[index]))
            flag = 1
        index = index+1
    if flag==1:
        bubblesort(tosort)
    return out
```

```python
tosort_start=[5,1,2,3,4,4,4,4,4,4,4]
```

```python
%timeit -n 10 bubblesort(tosort_start)
```

```python
# Same but counts iterations (slower for nearly-sorted lists)

def bubblesort_noshort(tosort, iteration=0):
    out = copy.copy(tosort)
    #for index, element in enumerate(tosort):
    index= 0
    flag = 0 #flags if an iteration does not require sorting
    while index+1 < len(tosort):
        #print({"list": tosort, "index": index, "iteration": iteration })
        if tosort[index] > tosort[index+1]: 
            tosort[index+1], tosort[index] = tosort[index], tosort[index+1]
            #print("flipped " + str(tosort[index]))
            #flag = 1
        index = index+1
    iteration += 1
    # can look for a flag to say it's sorted (flag=1) OR can count iterations and iterate n times
    if iteration < len(tosort) and flag == 1:
        bubblesort(tosort, iteration)
    else:
        return out
```

```python
tosort_start=[5,1,2,3,4]
```

```python
%timeit -n 10 bubblesort_noshort(tosort_start)
```
