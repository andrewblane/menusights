## Classification strategy

Training:

    Descriptors: Not extracted
    
    X: `culled_array`
    
    y: `y`
    
    
Test:

    Descriptors: `masked_ground_truth_x`
    
    X: `masked_ground_truth_vectorized`
    
    y: `masked_ground_truth_y`


```python
import numpy as np
def balanced_sample_maker(X, y, sample_size, random_seed=None):
    """ return a balanced data set by sampling all classes with sample_size 
        current version is developed on assumption that the positive
        class is the minority.

    Parameters:
    ===========
    X: {numpy.ndarrray}
    y: {numpy.ndarray}
    """
#     uniq_levels = np.unique(y)
#     uniq_counts = {level: sum(y == level) for level in uniq_levels}
    uniq_levels = list(Counter(y).keys())
    uniq_counts = list(Counter(y).values())
    
    if not random_seed is None:
        np.random.seed(random_seed)

    # find observation index of each class levels
    groupby_levels = {}
    for ii, level in enumerate(uniq_levels):
        obs_idx = [idx for idx, val in enumerate(y) if val == level]
        groupby_levels[level] = obs_idx
    #print(groupby_levels)
    # oversampling on observations of each label
    balanced_copy_idx = []
    for gb_level, gb_idx in groupby_levels.iteritems():
        over_sample_idx = np.random.choice(gb_idx, size=sample_size, replace=True).tolist()
        balanced_copy_idx+=over_sample_idx
    np.random.shuffle(balanced_copy_idx)
    #print(balanced_copy_idx)
    return (X[balanced_copy_idx, :], y[balanced_copy_idx], balanced_copy_idx)
```

```python
def evaluate_model(model):
    %matplotlib inline

    # predictions = logreg.predict(balanced_samples[0])
    # ground_truth = balanced_samples[1]
    predictions = [model.predict(i) for i in csr_matrix(balanced_samples[0]) if i.nnz>0]
    ground_truth = [balanced_samples[1][index] for index, i in enumerate(csr_matrix(balanced_samples[0])) if i.nnz>0]
    cm = confusion_matrix(ground_truth, predictions, labels=np.array(["low", "med", "high", "vhigh"]))

    #plt.figure(figsize = (10,7))
    print(model.score(np.array([i for i in balanced_samples[0] if csr_matrix(i).nnz>0]), goodindices))
    sns.heatmap(cm, annot=True, xticklabels=["low", "med", "high", "vhigh"], yticklabels=["low", "med", "high", "vhigh"])
```

```python
#iterate over values of C for logreg:
gridsearch = pd.DataFrame()
for C in np.logspace(-2,2,20):
    logreg = LogisticRegression(random_state=1323, penalty="l2", C=C, solver="lbfgs")
    logreg = logreg.fit(culled_array, y)
    score = logreg.score(np.array([i for i in balanced_samples[0] if csr_matrix(i).nnz>0]), goodindices)
    gridsearch = gridsearch.append({"C": C, "score": score}, ignore_index=True)
```
